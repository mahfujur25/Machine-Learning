```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Matching Learning Dataset/carprices.csv")
```

```python
df
```

```python
df['Car Model'].value_counts()
```

```python
dummies = pd.get_dummies(df['Car Model'])
```

```python
dummies
```

```python
merged = pd.concat([df,dummies], axis=1)
```

```python
merged
```

```python
final = merged.drop(["Car Model"], axis = 1)
```

```python
final
```

```python
y = df["Sell Price"]
```

```python
y
```

```python
x = final.drop(["Sell Price"], axis=1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
Accuracy = model.score(x,y)*100
```

```python
Accuracy
```
